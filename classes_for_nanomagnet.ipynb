```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

class NanoMagnet:
    def __init__(self, x, y, spin, image=None):
        self.x = x
        self.y = y
        self.spin = spin  # Spin values: 0: down, 1: right, 2: up, 3: left
        self.rotation, self.spin_value = self._get_rotation_and_spin(spin)
        self.image = image
        self.nearest_neighbor_list = []

    def _get_rotation_and_spin(self, spin):
        if spin == 0:
            return 90, 1
        elif spin == 1:
            return 0, 1
        elif spin == 2:
            return 90, -1
        elif spin == 3:
            return 0, -1

    def plot_nearest_neighbors(self):
        fig, ax = plt.subplots()
        if self.nearest_neighbor_list:
            x_pos = np.array(self.nearest_neighbor_list)[:, 0]
            y_pos = np.array(self.nearest_neighbor_list)[:, 1]
            ax.scatter(x_pos, y_pos, label="Nearest Neighbors")
        ax.scatter(self.x, self.y, color="red", label="Current Magnet")
        ax.set_aspect('equal')
        ax.legend()
        plt.show()

    def _get_nearest_neighbor_relative_coordinates(self):
        new_x_list, new_y_list = [], []
        for nearest_neighbor in self.nearest_neighbor_list:
            new_x_list.append(nearest_neighbor[0] - self.x)
            new_y_list.append(nearest_neighbor[1] - self.y)
        return new_x_list, new_y_list

class NanoMagnetArray:
    def __init__(self):
        self.nanomagnet_list = []

    @property
    def x_position(self):
        return np.array([nanomagnet.x for nanomagnet in self.nanomagnet_list])

    @property
    def y_position(self):
        return np.array([nanomagnet.y for nanomagnet in self.nanomagnet_list])

    def get_nanomagnet_position_list(self):
        return [[nanomagnet.x, nanomagnet.y] for nanomagnet in self.nanomagnet_list]

    def _get_all_nearest_neighbor_relative_coordinates(self):
        x_list, y_list = [], []
        for nanomagnet in self.nanomagnet_list:
            temp_x, temp_y = nanomagnet._get_nearest_neighbor_relative_coordinates()
            x_list.extend(temp_x)
            y_list.extend(temp_y)
        return x_list, y_list

    def find_nearest_neighbors(self, nearest_neighbors=9, leafsize=100):
        position_list = self.get_nanomagnet_position_list()
        nearest_neighbor_data = sp.spatial.cKDTree(position_list, leafsize=leafsize)
        for ip, nanomagnet in enumerate(self.nanomagnet_list):
            nn_data_list = nearest_neighbor_data.query(position_list[ip], nearest_neighbors)
            nn_link_list = []
            for nn_link in nn_data_list[1][1:]:  # Skipping the first element
                nn_link_list.append(position_list[nn_link])
            nanomagnet.nearest_neighbor_list = nn_link_list

    def find_nearest_neighbors_from_vertices(self, vertices_list):
        position_list = self.get_nanomagnet_position_list()
        nearest_neighbor_data = sp.spatial.cKDTree(position_list)
        vertex_neighbor_spins = {}
        for vertex in vertices_list:
            distances, indices = nearest_neighbor_data.query(vertex, k=4)
            nearest_neighbors = [self.nanomagnet_list[idx] for idx in indices]
            spins = [(nm.rotation, nm.spin_value) for nm in nearest_neighbors]
            vertex_neighbor_spins[tuple(vertex)] = spins
        return vertex_neighbor_spins

    def is_type_1_configuration(self, vertex, neighbors):
        # Separate the neighbors by rotation
        rotation_0 = [nm for nm in neighbors if nm.rotation == 0]
        rotation_90 = [nm for nm in neighbors if nm.rotation == 90]

        # Ensure we have the right number of neighbors
        if len(rotation_0) != 2 or len(rotation_90) != 2:
            return "Unknown"

        # Identify the nanomagnets by position
        low_x = min(rotation_0, key=lambda nm: nm.x)
        high_x = max(rotation_0, key=lambda nm: nm.x)
        high_y = max(rotation_90, key=lambda nm: nm.y)
        low_y = min(rotation_90, key=lambda nm: nm.y)

        # Check Type 1 configuration conditions
        type_1_condition_1 = (low_x.spin_value == -1 and high_x.spin_value == 1 and
                              high_y.spin_value == 1 and low_y.spin_value == -1)
        type_1_condition_2 = (low_x.spin_value == 1 and high_x.spin_value == -1 and
                              high_y.spin_value == -1 and low_y.spin_value == 1)
        
        # Check Type 2 configuration conditions
        type_2_condition_1 = (low_x.spin_value == -1 and high_x.spin_value == -1 and
                              high_y.spin_value == 1 and low_y.spin_value == 1)
        type_2_condition_2 = (low_x.spin_value == -1 and high_x.spin_value == -1 and
                              high_y.spin_value == -1 and low_y.spin_value == -1)
        type_2_condition_3 = (low_x.spin_value == 1 and high_x.spin_value == 1 and
                              high_y.spin_value == 1 and low_y.spin_value == 1)
        type_2_condition_4 = (low_x.spin_value == 1 and high_x.spin_value == 1 and
                              high_y.spin_value == -1 and low_y.spin_value == -1)
        
        
        # Check Type 4 configuration conditions
        type_4_condition_1 = (low_x.spin_value == 1 and high_x.spin_value == -1 and
                              high_y.spin_value == 1 and low_y.spin_value == -1)
        type_4_condition_2 = (low_x.spin_value == -1 and high_x.spin_value == 1 and
                              high_y.spin_value == -1 and low_y.spin_value == 1)

        if (type_1_condition_1 or type_1_condition_2):
            return "Type 1"
        elif (type_2_condition_1 or type_2_condition_2 or type_2_condition_3 or type_2_condition_4):
            return "Type 2"
        elif (type_4_condition_1 or type_4_condition_2):
            return "Type 4"
        else:
            return "Type 3"

    def find_and_classify_spin_configurations(self, vertices_list):
        position_list = self.get_nanomagnet_position_list()
        nearest_neighbor_data = sp.spatial.cKDTree(position_list)
        classified_configurations = {}

        for vertex in vertices_list:
            distances, indices = nearest_neighbor_data.query(vertex, k=4)
            nearest_neighbors = [self.nanomagnet_list[idx] for idx in indices]
            configuration = self.is_type_1_configuration(vertex, nearest_neighbors)
            classified_configurations[tuple(vertex)] = configuration

        return classified_configurations

    def count_spin_configurations(self, vertices_list):
        classified_configurations = self.find_and_classify_spin_configurations(vertices_list)
        configuration_counts = {'Type 1': 0, 'Type 2': 0, 'Type 3': 0, 'Type 4': 0, 'Unknown': 0}

        for configuration in classified_configurations.values():
            if configuration in configuration_counts:
                configuration_counts[configuration] += 1
            else:
                configuration_counts['Unknown'] += 1

        return configuration_counts

    def plot_nearest_neighbors_of_vertices(self, vertices_list):
        position_list = self.get_nanomagnet_position_list()
        nearest_neighbor_data = sp.spatial.cKDTree(position_list)
        for vertex in vertices_list:
            distances, indices = nearest_neighbor_data.query(vertex, k=4)
            nearest_neighbors = [position_list[idx] for idx in indices]
            fig, ax = plt.subplots()
            vertex_x, vertex_y = vertex
            nn_x = [nn[0] for nn in nearest_neighbors]
            nn_y = [nn[1] for nn in nearest_neighbors]
            ax.scatter(nn_x, nn_y, label="Nearest Neighbors")
            ax.scatter(vertex_x, vertex_y, color="red", label="Vertex")
            ax.set_aspect('equal')
            ax.legend()
            plt.show()
```
